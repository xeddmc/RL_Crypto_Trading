```python
import pandas as pd
import numpy as np
import os
from prophet import Prophet
import plotly.graph_objects as go
from sklearn.metrics import r2_score # R^2 score
from sklearn.metrics import mean_squared_error # squared = True for MSE, False for RMSE
from sklearn.metrics import mean_absolute_error # mean absolute error
import pandas as pd
import numpy as np
import os
from prophet import Prophet
import plotly.graph_objects as go
from sklearn.metrics import r2_score # R^2 score
from sklearn.metrics import mean_squared_error # squared = True for MSE, False for RMSE
from sklearn.metrics import mean_absolute_error # mean absolute error
```

# Fbprophet in multiplicative mode
Plots of forecasts on 7 days and regression metrics

## 1 day data

```python
day_dir = '../../data/ready/1-day-data/{}'
day_files = os.listdir(day_dir.format(''))
```

Description of plots:

black dots: historical values

blue line: forecasted values

light blue area: uncertainty intervals 

```python
# help(Prophet)
```

```python
for filename in day_files:
    df = pd.read_csv(day_dir.format(filename))
    df['y'] = df['close']
    m = Prophet(seasonality_mode='multiplicative')
    
    m.fit(df[:-7])
    
    future = m.make_future_dataframe(periods=7)
    forecast = m.predict(future)
    
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=df['ds'][0:-7], y=df['y'][0:-7], name='Train'))
    
    fig.add_trace(go.Scatter(x=forecast['ds'], y=forecast['yhat'], name='Forecasts'))
    
    fig.add_trace(go.Scatter(x=df['ds'][-7:], y=df['y'][-7:], name='Actual'))
    fig.show()
    # fig1 = m.plot(forecast)
    r2 = r2_score(df['y'][-7:],forecast['yhat'][-7:])
    mse = mean_squared_error(df['y'][-7:], forecast['yhat'][-7:])
    rmse = mean_squared_error(df['y'][-7:], forecast['yhat'][-7:], squared=False)
    mae = mean_absolute_error(df['y'][-7:], forecast['yhat'][-7:])
    print('{}\nR2: {}\nMSE: {}\nRMSE: {}\nMAE: {}\n'.format(filename, r2, mse, rmse, mae))
```

## ARIMA 
endog = dependent variable

exog = independent variable

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
for filename in day_files:
    df = pd.read_csv(day_dir.format(filename), index_col=0, parse_dates=True)
    model = ARIMA(df['close'][:-7], order=(5,2,5), freq='D').fit()
    yhat = model.forecast(steps=7)
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=df.index[:-7], y=df['close'][:-7], name='Train'))
    fig.add_trace(go.Scatter(x=df.index[-7:], y=yhat, name='Forecasts'))
    fig.add_trace(go.Scatter(x=df.index[-7:], y=df['close'][-7:], name='Actual'))
    fig.show()

    r2 = r2_score(df['close'][-7:],yhat)
    mse = mean_squared_error(df['close'][-7:], yhat)
    rmse = mean_squared_error(df['close'][-7:], yhat, squared=False)
    mae = mean_absolute_error(df['close'][-7:], yhat)
    print('{}\nR2: {}\nMSE: {}\nRMSE: {}\nMAE: {}\n'.format(filename, r2, mse, rmse, mae))

```

## Holt-Winters

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
for filename in day_files:
    df = pd.read_csv(day_dir.format(filename), index_col=0, parse_dates=True)
    df.index.freq = 'D'
    model = ExponentialSmoothing(df['close'][:-7], trend='mul').fit()
    yhat = model.forecast(7)
    
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=df.index[:-7], y=df['close'][:-7], name='Train'))
    fig.add_trace(go.Scatter(x=df.index[-7:], y=yhat, name='Forecasts'))
    fig.add_trace(go.Scatter(x=df.index[-7:], y=df['close'][-7:], name='Actual'))
    fig.show()

    r2 = r2_score(df['close'][-7:],yhat)
    mse = mean_squared_error(df['close'][-7:], yhat)
    rmse = mean_squared_error(df['close'][-7:], yhat, squared=False)
    mae = mean_absolute_error(df['close'][-7:], yhat)
    print('{}\nR2: {}\nMSE: {}\nRMSE: {}\nMAE: {}\n'.format(filename, r2, mse, rmse, mae))
```
